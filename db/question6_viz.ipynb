```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('question6(wines).csv')
df.groupby('country_name')['wines_per_country'].sum().sort_values(ascending=False)
```

```python
df.columns.tolist()
```

```python
palette = sns.color_palette("plasma", as_cmap=True)


plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df, 
    x='country_name', 
    y='weighted_average_rating', 
    size='wines_per_country', 
    hue='wines_per_country', 
    palette=palette, 
    sizes=(400, 1200), 
    legend='brief')


plt.xlabel('Names of countries')
plt.ylabel('Average rating per country')
plt.title('Average wine per Country (Wine table)')
plt.xticks(rotation=45, ha='right')

```

```python
df2 = pd.read_csv('question6(vintage).csv')
df2_filtered = df2[df2['country_name'] != 'Israël']
```

```python

palette = sns.color_palette("plasma", as_cmap=True)


plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df2_filtered, 
    x='country_name', 
    y='weighted_average_rating', 
    size='wines_per_country', 
    hue='wines_per_country', 
    palette=palette, 
    sizes=(400, 1200), 
    legend='brief')


plt.xlabel('Names of countries')
plt.ylabel('Average rating per country')
plt.title('Average wine per Country (Vintage table)')
plt.xticks(rotation=45, ha='right')

```
